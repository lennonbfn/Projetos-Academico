# **2.   Importar imagem que contém o padrão a ser encontrado**



```python
image1 = cv2.cvtColor(cv2.imread('/content/coke_bottle.png'),cv2.COLOR_BGR2RGB)
template1 =  cv2.cvtColor(cv2.imread('/content/coke_logo.png'),cv2.COLOR_BGR2RGB)
image2 =  cv2.cvtColor(cv2.imread('/content/8_diamonds.png'),cv2.COLOR_BGR2RGB)
template2 =  cv2.cvtColor(cv2.imread('/content/diamonds_template.png'),cv2.COLOR_BGR2RGB)
#new_image = ((cv2.threshold(img,127,255,cv2.THRESH_BINARY)[1])/255)

plt.imshow(template1)
```

```python
imageGray = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
templateGray = cv2.cvtColor(template1, cv2.COLOR_BGR2GRAY)
imageGray2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)
templateGray2 = cv2.cvtColor(template2, cv2.COLOR_BGR2GRAY)
```

```python
print("[INFO] performing template matching...")
result1 = cv2.matchTemplate(imageGray, templateGray,cv2.TM_CCOEFF_NORMED)
(minVal1, maxVal1, minLoc1, maxLoc1) = cv2.minMaxLoc(result1)
```

```python
print("[INFO] performing template matching...")
result2 = cv2.matchTemplate(imageGray2, templateGray2,cv2.TM_CCOEFF_NORMED)
(minVal2, maxVal2, minLoc2, maxLoc2) = cv2.minMaxLoc(result2)
print()
```

```python
(startX, startY) = maxLoc1
endX = startX + template1.shape[1]
endY = startY + template1.shape[0]
```

```python
(startX2, startY2) = maxLoc2
endX2 = startX2 + template2.shape[1]
endY2 = startY2 + template2.shape[0]
```

```python
a = cv2.rectangle(image1, (startX, startY), (endX, endY), (0, 0, 0), 3)
plt.imshow(a)
```

```python
b = cv2.rectangle(image2, (startX2, startY2), (endX2, endY2), (0, 0, 0), 3)
plt.imshow(b)
```
